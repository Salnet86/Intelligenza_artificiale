```python
df2=pd.read_csv('')
```

```python
df2
```

```python
df2['TMEDIA °C'] = df2['TMEDIA °C'].str.replace(',', '.').astype(float)
```

```python
print(df2['TMEDIA °C'].dtype)
print(df2['TMEDIA °C'].head())
```

```python
import matplotlib.pyplot as plt

df2['TMEDIA °C'].plot(kind='line')
plt.xlabel('Indice')
plt.ylabel('Temperatura media (°C)')
plt.title('Grafico a barre della temperatura media')
plt.show()
```

```python
df2['TMIN °C'].info()
```

```python
import matplotlib.pyplot as plt

df2['TMIN °C'].plot(kind='line', color='skyblue')
plt.xlabel('indice')
plt.ylabel('Temperatura minima (°C)')
plt.title('Temperatura minima giornaliera')
plt.show()
```

```python
df2['TMAX °C']=df2['TMAX °C'].str.replace(',', '.').astype(float)
```

```python
df2['TMAX °C'].info()
```

```python
import matplotlib.pyplot as plt

df2['TMAX °C'].plot(kind='line', color='skyblue')
plt.xlabel('indice')
plt.ylabel('Temperatura massima (°C)')
plt.title('Temperatura minima giornaliera')
plt.show()
```
