```python
x=range(len(df["ID"]))
y=df["TMEDIA °C"]

```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, marker="o")
plt.xlabel("Giorno di Giugno")
plt.ylabel("Temperatura media (°C)")
plt.title("Temperatura media - Giugno 2010")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
